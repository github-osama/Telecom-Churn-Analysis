# **Introduction about the datafile**

**Industry Introduction**

In the telecom sector, the telecom company works to make
communication possible on a global scale. The telecom company provides services to the customer in the form of the phone call, text messages, voicemails, and internet service. At the current time, the telecom sector is a rapidly growing sector, but the customers of this sector are not loyal for only one service provider or company. The reason is that in this industry lots of companies are operating due to this customers are able to choose services from more than one company or service provider. The customer of this sector actively switches from one service provider to the other service provider. Because of this reason, this industry experiences a huge amount of **churn or less retention rate of the customer**. The average churn rate of the customer in this sector (telecom sector) can be found anywhere between 15 to 25 percent.

In this sector, it is fact that the companies invest 5 to 10 times more money to acquire a new customer from the industry as compared to the money invested to retain the existing customer. So, customer retention is highly important in this sector for the growth of a company.

For a telecom service provider company, decrease in customer retention (or churn) is a big problem. So, the service provider company should work to find the **factors which is causing the churn** so that they can use those factors and improve quality. 

**Objective of this exploratory data analysis (EDA)**

In this project we will to find the factors which causes the churn. So, that the service provider can use those factors to decrese the churn.

**Introduction of the given dataset**

The given data file contains customer details of a French multinational telecommunications corporation. The name of the telecommunications corporation **Orange S.A., formerly France Télécom S.A**. This data file contains 3333 customers, and the data file is given in the form of a CSV (comma separated file) file.

**The dataset contains the details of following columns:**

**STATE:** Contains a short form of the 51 Unique States' name.

**Account Length:** Represents the Length of The Account (in terms of usage).

**Area Code:** Represents the Code number of the area having the same States.

**International Plan:** (contains Yes/No) Yes represents that the customer is using an international Plan and No represents that the customer is not using an international Plan.

**Voice Mail Plan:** (contains Yes/No) Yes represents that the customer is using a voicemail plan and No represents that the customer is not using a voicemail plan.

**Number vmail messages:** Represents the number of voice mail messages used by the customer.

**Total day minutes:** This represents the total number of minutes a customer spends in the day (morning) time.

**Total day calls:** Represents the total number of calls made by the customer in the morning time.

**Total day charge:** Represents the total amount charged by the company from the customers in the morning time.

**Total eve minutes:** Represents the total number of minutes a customer spends in the evening time.

**Total eve calls:** Represents the total number of calls made by the customer in the evening.

**Total eve charge:** Represents the total amount charged by the company from the customers in the evening time.

**Total night minutes:** This represents the total number of minutes a customer spends at night time.

**Total night calls:** Represents the total number of calls made by the customer in the night evening.

**Total night charge:** Represents the total amount charged by the company from the customers in the night-time.

**Total intl minutes:** This represents the total number of minutes a customer spends on international calls.

**Total intl calls:** Represents the total number of international calls made by the customer.

**Total intl charge:** Represents the total amount charged by the company from the customers in the international calls.

**Customer service calls:** Number of customer service calls made by the customer.

**Churn Customer Churn:** (contains True/False) True means churned customer, False means retained customer

**(6) Churn of customer according to international plan**

```python
#Check churn and non churn of customer using the international plan
#Dataframe of churn according to international plan
xy_int = dataframe.groupby(['International plan'])['Churn'].value_counts().unstack().reset_index()
xy_int = xy_int.rename(columns = {False:'Non Churn', True: 'Churn'})

#Function to calculate summation of two number
def colsum(x,y):
  '''It will take column as input'''
  return x + y

#Add total column in the xy_int dataframe
xy_int['Total'] = xy_int.apply(lambda x: colsum(x['Non Churn'], x['Churn']), axis = 1)

#Function to calculate percentage
def calcpercent(x,y):
  '''It will take column as input'''
  return (x/y)*100

#Add percentage column in the xy_int dataframe
xy_int['Percentage_churn'] = xy_int.apply(lambda x: round(calcpercent(x['Churn'], x['Total']),1), axis = 1)
xy_int['Percentage_non_churn'] = xy_int.apply(lambda x: round(100 - x['Percentage_churn'],1), axis = 1)

#Shows the churn and non churn customers who is using and not using international plan
sns.countplot(x = 'International plan', hue = 'Churn', data = dataframe)
plt.xlabel('Churn in International plan')
plt.ylabel('Number of customers')
plt.show()
print('\n')

xy_int
```

Here, we can see that 11.5% of customer not using internaitonal plan is churned and 88.5% is not churned.

And 42.4% of customer using international plan is churned and 57.6% are not churned. It means that approxmaitely half of the customer using International plan is churned.

It means that the churn rate of the customer using internaitonal plan is high as compared to the customer who is not using international plan.

**(11) Check Churn in customer according to the mean price charged by the company in day, evening, night, and international calls**

```python
#Calcualte mean price charged by the customer according to usage: day, night, eve (evening), and intl (international)
#Mean price (Day usage)
mean_price_day = round(dataframe['Total day charge'].mean()/dataframe['Total day minutes'].mean(), 2)
print(f'The average price charged by the customer in day usage is: {mean_price_day}')

#Mean price (Evening usage)
mean_price_eve = round(dataframe['Total eve charge'].mean()/dataframe['Total eve minutes'].mean(),2)
print(f'The average price charged by the customer in evening usage is: {mean_price_eve}')

#Mean price (Night usage)
mean_price_night = round(dataframe['Total night charge'].mean()/dataframe['Total night minutes'].mean(),2)
print(f'The average price charged by the customer in night usage is: {mean_price_night}')

#Mean price (International usage)
mean_price_intl = round(dataframe['Total intl charge'].mean()/dataframe['Total intl minutes'].mean(),2)
print(f'The average price charged by the customer in international usage is: {mean_price_intl}')
print('\n')

#Show the usage in the bargraph
plt.figure(figsize=(7,7))
sns.barplot(x = ['Day Call', 'Evening Call', 'Night Call', 'International Call'], y = [mean_price_day, mean_price_eve, mean_price_night, mean_price_intl])
plt.title('Mean price per minute comparison price')
plt.ylabel('Mean price per minute')
plt.show()
```

Here, we can see that the per minute price charged by the user in International call and day call is higher as comapred to evening and night call.
